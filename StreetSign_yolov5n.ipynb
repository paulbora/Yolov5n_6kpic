```python
import torch
import cv2
from PIL import Image
import numpy as np

model_path = "C:\\Users\\paul.bora\\Desktop\\WorkInProgress\\Yolov5n\\yolov5\\runs\\train\\yolov5n_results4\\weights\\last.pt"
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

video_path = "C:\\Users\\paul.bora\\Desktop\\video_semn_trafic\\PXL_20231028_124213967.TS.mp4"
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, (640, 480))

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    img = Image.fromarray(rgb_frame)

    results = model(img)

    output_frame = results.render()[0]

    output_frame = cv2.cvtColor(np.array(output_frame), cv2.COLOR_RGB2BGR)

    cv2.imshow('Traffic Sign', output_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()

cv2.destroyAllWindows()
```
